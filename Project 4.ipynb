```python
import folium
import requests
import pandas

arrest_table = pandas.read_csv("http://www.hcbravo.org/IntroDataSci/misc/BPD_Arrests.csv")

arrest_table["race_new"] = arrest_table["sex"]
arrest_table["sex_new"] = arrest_table["race"]
arrest_table["race"] = arrest_table["race_new"]
arrest_table["sex"] = arrest_table["sex_new"]
arrest_table = arrest_table.drop('race_new', 1)
arrest_table = arrest_table.drop('sex_new', 1)

arrest_table = arrest_table[pandas.notnull(arrest_table["Location 1"])]

arrest_table["lat"], arrest_table["long"] = arrest_table["Location 1"].str.split(",").str
arrest_table["lat"] = arrest_table["lat"].str.replace("(", "").astype(float)
arrest_table["long"] = arrest_table["long"].str.replace(")", "").astype(float)

map_osm = folium.Map(location=[39.29, -76.61], zoom_start=11)

color_dict = {'M': 'red', 'F': 'blue'}

i=0
for index, row in arrest_table.iterrows():
    if row.chargeDescription == 'Unknown Charge':
        continue
    lat = row.lat
    long = row.long
    sex = row.sex
    folium.CircleMarker([lat, long], radius=5, popup=str(row.chargeDescription), color=color_dict[sex],fill=True, fill_color=color_dict[sex]).add_to(map_osm)
    i += 1
    if i == 1000:
        break

map_osm
```

What I am showing with this map is the first 1000 cases of the arrest_table dataframe as circle markers. I've color coded the markers as Red for Male and Blue for Female arrestees. I've chosen to do only the first 1000 cases because the amount of data provided was too time consuming for the program to be run and also created a map that was too chaotic. By chaotic, I mean that there were too many datapoints to be seen all at once. With 1000 datapoints, it is possible to zoom out and see the entire city and easily discern where crimes are concentrated and allows zooming in to give a good separation of crimes (fewer marker overlap). I've also included a popup for each marker to display the charge of the crime.

With this map we can see that:
1. Most crimes are commited by males.
2. There is an isolated area in the upper western region concentrated with crimes related to drugs.
3. The majority of crimes occur in the city near the harbor, North of I-95.
4. South of I-95, crimes are most likely to occur near the river
